```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = './/transfer//1702621_376_26624_29696.jpg'
image = cv2.imread(path)
plt.imshow(image)
```

```python
B,G,R = cv2.split(image)
plt.imshow(B)
```

```python
plt.imshow(G)
```

```python
plt.imshow(R)
```

```python
ret, th = cv2.threshold(G, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
ret
plt.imshow(th)
```

```python
print(ret)
```

```python
ret_R, th_R = cv2.threshold(R, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(ret_R)
plt.imshow(th_R)
```

```python
ret_B, th_B = cv2.threshold(B, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(ret_B)
plt.imshow(th_B)
```

```python
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(image_hsv)
```

```python
ret_H, th_H = cv2.threshold(H, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(ret_H)
plt.imshow(th_H)
```

```python
ret_S, th_S = cv2.threshold(S, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(ret_S)
plt.imshow(th_S)
```

```python
ret_V, th_V = cv2.threshold(V, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(ret_V)
plt.imshow(th_V)
```
